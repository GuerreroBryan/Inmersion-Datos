```python
# 2
plt.figure(figsize=(12,8))

inmuebles["Precio_m2"] = (inmuebles.Precio_Millones / inmuebles.Area)
inmuebles.info()
```

```python
precio_m2_barrio = pd.DataFrame(inmuebles.groupby("Barrio")["Precio_m2"].mean().sort_values(ascending=False))
precio_m2_barrio.reset_index(inplace=True)
print(f"El dataset contiene {len(precio_m2_barrio)} barrios")

```

```python
grafica = sns.barplot(x="Precio_m2", y="Barrio", palette="viridis", data=precio_m2_barrio.head(10))

grafica.set_title("Precio promedio por metro cuadrado, por barrio (Top 10)", fontsize=14)
grafica.set_xlabel("Precio promedio por m2 (millones)", fontsize=13)
grafica.set_ylabel("Barrio", fontsize=13)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.savefig("dia2_desafio2.png", format="png")

plt.show()
```
